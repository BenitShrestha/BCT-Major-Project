```python
import pandas as pd
```

```python
df = pd.read_csv('clean_date_categories.csv',encoding='utf-8-sig')
df.head()
```

```python
df.columns
```

```python
new_df = df[['text']]
new_df.head()
```

NAN VALUES

```python
nan_count = new_df['text'].isna().sum()
print(f"Number of NaN values in 'text' column: {nan_count}")
```

```python
print(new_df[new_df['text'].isna()])
```

```python
# Drop rows with NaN in 'text'
new_df = new_df.dropna(subset=['text'])
```

```python
nan_count = new_df['text'].isna().sum()
print(f"Number of NaN values in 'text' column: {nan_count}")
```

DUPLICATE VALUES

```python
duplicate_count = new_df['text'].duplicated().sum()
print(f"Number of duplicate values in 'text' column: {duplicate_count}")
```

```python
import re

def clean_text(text):
    text = text.strip()
    # Remove place name patterns (typically up to first "।")
    text = re.sub(r'^.*?।', '', text, 1)  # Remove up to first "।"
    return text.strip()

new_df['cleaned_text'] = new_df['text'].apply(clean_text)
```

```python
new_df = new_df['cleaned_text']
```

```python
new_df
```

```python
# Save CSV with UTF-8 encoding
new_df.to_csv('nepberta_dataset.csv', index=False, encoding='utf-8-sig') 
```

```python
df = pd.read_csv('nepberta_dataset.csv',encoding='utf-8-sig')
df.head()
```

```python
with open("nepberta_text.txt", "w", encoding="utf-8") as f:
    for line in df['cleaned_text']:
        if isinstance(line, str) and line.strip():  # write only non-empty strings
            f.write(line.strip() + "\n")
```
